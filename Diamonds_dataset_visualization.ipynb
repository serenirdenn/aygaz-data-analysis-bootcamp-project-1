```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pandas.api.types import CategoricalDtype
import seaborn as sns

sns.set(rc={"figure.figsize":(6, 4)})
```

```python
df = pd.read_csv("diamonds.csv", delimiter=",", encoding="UTF-8")
df.head()
```

```python
df["cut"].index
```

```python
sns.barplot(x= "cut", y=df["cut"].index, data=df,color="purple").set_title("Cut Density")
```

```python
sns.barplot(x= "cut", y="price", hue="color",data=df)
plt.title("Cut - Price - Color")
```

```python
sns.barplot(x="cut", 
            y="price", 
            hue="clarity",
            data=df)
plt.title("Cut - Price - Clarity")
```

```python
sns.barplot(x="cut", 
            y="price", 
            hue="carat",
            data=df)
plt.title("Cut - Price - Carat")
```

```python
df.info()
```

```python
df.groupby("cut").sum().plot.pie(y= "price", autopct="%.1f%%", ylabel= " ", legend=False, figsize=(5,5))
```

```python
cols=["x", "y", "z"]
for col in cols:
    plt.figure(figsize=(5,5))
    sns.scatterplot(x=col, y="price", data=df)
```
